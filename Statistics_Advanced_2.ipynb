1.  What is hypothesis testing in statistics?
    - Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample of data to support a specific claim about a population. It essentially tests whether an observed result is statistically significant or simply due to random chance.
    - I. The Core LogicThe process relies on two opposing statements:Null Hypothesis ($H_0$): The "no-effect" assumption. It claims there is no difference or change (e.g., "This drug does not work").Alternative Hypothesis ($H_a$): The claim you want to prove. It suggests there is a real effect or difference (e.g., "This drug does work").
    - II. Making the DecisionTo decide between these two, you calculate a P-value from your data:If the P-value is low (typically $\leq 0.05$): The result is unlikely to have happened by luck. You reject the Null Hypothesis in favor of the Alternative.If the P-value is high: The result could easily be a fluke. You fail to reject the Null Hypothesis.
    - III. Summary of StepsState your hypotheses ($H_0$ vs. $H_a$).Set a significance level (usually $\alpha = 0.05$).Collect data and calculate a test statistic.Compare the P-value to your significance level to reach a conclusion.

2.   What is the null hypothesis, and how does it differ from the alternative
hypothesis?
     - The Null Hypothesis ($H_0$) and the Alternative Hypothesis ($H_a$) are two opposing statements that cover all possible outcomes of a statistical test.
     - The Null Hypothesis ($H_0$) is the starting assumption that there is no effect, no relationship, or no difference between groups. It represents the "status quo" and suggests that any variation you see in your data is simply due to random chance. For example, if you are testing a new medicine, the null hypothesis would state that the medicine has no effect on patients compared to a placebo.
     - The Alternative Hypothesis ($H_a$), on the other hand, is the statement you are trying to find evidence for. It claims that there is a real effect, a significant difference, or a specific relationship present. In the medicine example, the alternative hypothesis would state that the medicine does indeed improve patient health.
     - The primary difference lies in the burden of proof. In statistics, we assume the Null Hypothesis is true by default—much like a defendant is "innocent until proven guilty." We only reject the Null Hypothesis and accept the Alternative Hypothesis if our data provides strong enough evidence (usually indicated by a low P-value) to show that the result is very unlikely to have happened by luck alone.

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
   - The significance level, denoted by the Greek letter alpha ($\alpha$), is a fixed threshold that determines how much evidence you need before you reject the null hypothesis. It represents the probability of making a "Type I error," which is the risk of concluding an effect exists when it actually doesn't.- Its Role in the ProcessBefore you even look at your data, you choose a significance level (commonly 0.05 or 5%). This acts as the "cutoff line" for your decision-making.
   - When you run your statistical test, you calculate a P-value, which tells you the probability that your results happened by random chance. You then compare that P-value to your significance level to decide the outcome:
   - If P-value $\leq \alpha$: The result is "statistically significant." You have enough evidence to reject the null hypothesis and support your claim.
   - If P-value $> \alpha$: The result is not statistically significant. You do not have enough evidence to reject the null hypothesis, meaning the outcome could likely be due to chance.
   - Why It MattersThe significance level essentially defines your standard of proof. A smaller $\alpha$ (like 0.01) makes it much harder to claim a discovery, requiring very strong evidence. A larger $\alpha$ (like 0.10) is more lenient but increases the risk of being wrong and claiming an effect that isn't really there.

4. What are Type I and Type II errors? Give examples of each.
   - In hypothesis testing, because we are using samples to make guesses about whole populations, there is always a small chance we will reach the wrong conclusion. These mistakes are categorized as Type I and Type II errors.
   - Type I Error (False Positive)A Type I error occurs when you reject the null hypothesis even though it is actually true. In other words, you conclude that an effect or difference exists when, in reality, it does not. The probability of committing this error is equal to your significance level ($\alpha$).
   - Example: A medical test indicates that a patient has a specific disease when they are actually healthy.
   - The Mistake: Finding an effect that isn't there.
   - Type II Error (False Negative)A Type II error occurs when you fail to reject the null hypothesis even though it is actually false. This means you conclude there is no effect or difference, when in fact there is one. The probability of committing this error is denoted by the Greek letter beta ($\beta$).
   - Example: A medical test results in a "clear" or negative reading for a patient who actually has the disease.
   - The Mistake: Failing to find an effect that is actually there.

5.  What is the difference between a Z-test and a T-test? Explain when to use
each.
    - The main difference between a Z-test and a T-test lies in how much information you have about the population you are studying, specifically regarding the standard deviation and the sample size.
    - The Z-TestThe Z-test is used when you are dealing with large samples or when the population parameters are already known.
    - When to use it: You use a Z-test when the sample size is large (typically $n \geq 30$) and the population standard deviation is known. Because the sample is large, the central limit theorem allows us to assume a normal distribution.
    - Example: Testing the average height of adult men in a country where the national standard deviation for height is already well-documented from census data.
    - The T-TestThe T-test is used when data is more limited or the population's true variability is a mystery.
    - When to use it: You use a T-test when the sample size is small (typically $n < 30$) and the population standard deviation is unknown. In this case, you must estimate the standard deviation using your sample data. The T-test uses a "t-distribution," which has thicker tails to account for the extra uncertainty of a small sample.
    - Example: Testing the effectiveness of a new drug in a small clinical trial with only 15 participants.
    - Key Summary for ChoosingTo decide which test to use, ask two questions:
    - Do I know the population standard deviation? If yes, use a Z-test. If no, move to the second question.
    - Is my sample size large ($n \geq 30$)? If yes, you can technically use a Z-test (though a T-test still works). If the sample is small ($n < 30$), you must use a T-test.

```python
'''8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)'''
'''Interpretation
Sample Mean (48.96): This is the average of our 100 generated points.

Confidence Interval (47.16, 50.76): We are 95% confident that the true population mean lies between these two values. Since we originally set the population mean to 50 when generating the data, our interval successfully captures it.

The Plot: The histogram shows the spread of the data, while the dashed red line marks the average and the dotted blue lines show the boundaries of our confidence range.'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate 100 random samples from a normal distribution (mean=50, std=10)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate sample statistics
mean = np.mean(data)
std_err = stats.sem(data)  # Standard error of the mean
confidence_level = 0.95
df = len(data) - 1  # Degrees of freedom

# Calculate the 95% Confidence Interval
ci = stats.t.interval(confidence_level, df, mean, std_err)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, color='lightgreen', edgecolor='black', alpha=0.7)

# Add vertical lines for mean and CI boundaries
plt.axvline(mean, color='red', linestyle='dashed', label=f'Mean: {mean:.2f}')
plt.axvline(ci[0], color='blue', linestyle='dotted', label='95% CI Lower')
plt.axvline(ci[1], color='blue', linestyle='dotted', label='95% CI Upper')

plt.title('Normal Distribution with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
'''9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)'''
'''A Z-score (or standard score) represents how many standard deviations a specific data point is from the mean of the dataset.

Positive Z-score: The data point is above the mean.

Negative Z-score: The data point is below the mean.

Zero Z-score: The data point is exactly equal to the mean.

Standardizing data into Z-scores allows researchers to compare scores from different distributions or scales by bringing them onto a common scale with a mean of 0 and a standard deviation of 1.'''
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores: (x - mean) / std_dev
    z_scores = (data - mean) / std_dev

    # Visualization
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=15, color='orchid', edgecolor='black', alpha=0.7)

    # Add a vertical line at the mean (Z=0)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')

    plt.title('Histogram of Standardized Data (Z-scores)')
    plt.xlabel('Z-score (Standard Deviations from Mean)')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.show()

    return z_scores

# Generate sample dataset
np.random.seed(10)
sample_dataset = np.random.normal(loc=100, scale=15, size=500)

# Run function
z_results = calculate_and_plot_zscores(sample_dataset)

# Show first 5 Z-scores as an example
print("First 5 Z-scores:", z_results[:5])
```
